```python
import serial
import mysql.connector
from datetime import datetime
```

```python
#configuramos el puerto serial
arduino_port = "COM7"
baud_rate = 9600
```

```python
#configuracion mysql
db_host = "localhost"
db_user = "root"
db_password = "Einstein1729"
db_name = "sensordht11"
```

```python
#establecemos la conexion serial
ser = serial.Serial(arduino_port, baud_rate, timeout=1)
print("Connected to Arduino on port:", arduino_port)
```

```python
#Establecemos conexion serial
db = mysql.connector.connect(
    host=db_host,
    user=db_user,
    password=db_password,
    database=db_name
)
cursor = db.cursor()
print("Connected to MySQL database:", db_name)
```

```python
def insert_data(temperature, humidity):
    query = "INSERT INTO datos_sensor (temperatura, humedad) VALUES (%s, %s)"
    values = (temperature, humidity)
    cursor.execute(query, values)
    db.commit()
    print("Data inserted:", values)
    
try:
    while True:
        line = ser.readline().decode('utf-8').strip()
        if line:
            print("Received from Arduino:", line)
            if "Temperature" in line and "Humidity" in line:
                parts = line.split(",")
                temp_str = parts[0].split(":")[1].strip().replace("*C", "")
                hum_str = parts[1].split(":")[1].strip().replace("%", "")
                temperature = float(temp_str)
                humidity = float(hum_str)
                insert_data(temperature, humidity)
except KeyboardInterrupt:
    print("Program terminated")

finally:
    cursor.close()
    db.close()
    ser.close()
    print("Connections closed") 
```
